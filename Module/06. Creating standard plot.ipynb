### Creating standard plots (line,bar,pie)

```python
import numpy as np
from numpy.random import randn
import pandas as pd

from pandas import Series, DataFrame
import matplotlib.pyplot as plt

from matplotlib import rcParams

import seaborn as sb
```

```python
# set_style = whitegrid, dark, white, ticks
%matplotlib inline

rcParams['figure.figsize']= 5,4

sb.set_style('whitegrid')
```

### Creating aline chart from a list object

#### Plotting a line chart in matplotlib

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
plt.show()
```

#### Plotting a line chart from a pandas object

```python
address = 'F:\Learn R\Dataset\mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars['mpg']

mpg.plot()
plt.show()
```

```python
df = cars[['cyl','wt','mpg']]
df.plot()
plt.show()
```

### Creating bar charts

#### Creating a bar chart from a list

```python
plt.bar(x,y)
plt.show()
```

#### Creating bar chart from pandas objects

```python
mpg.plot(kind = 'bar')
plt.show()
```

```python
mpg.plot(kind = 'barh')
```

#### Creating a pie chart

```python
x = [1,2,3,4,5]
plt.pie(x)
plt.show()
```
